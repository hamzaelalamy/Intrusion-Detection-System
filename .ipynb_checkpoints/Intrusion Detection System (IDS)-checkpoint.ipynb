```python
# Importation des bibliothèques nécessaires
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix,classification_report
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
# Définition des noms de colonnes pour le jeu de données
columns=["duration",
 "protocol_type",
 "service",
 "flag",
 "src_bytes",
 "dst_bytes",
 "land",
 "wrong_fragment",
 "urgent",
 "hot",
 "num_failed_logins",
 "logged_in",
 "num_compromised",
 "root_shell",
 "su_attempted",
 "num_root",
 "num_file_creations",
 "num_shells",
 "num_access_files",
 "num_outbound_cmds",
 "is_host_login",
 "is_guest_login",
 "count",
 "srv_count",
 "serror_rate",
 "srv_serror_rate",
 "rerror_rate",
 "srv_rerror_rate",
 "same_srv_rate",
 "diff_srv_rate",
 "srv_diff_host_rate",
 "dst_host_count",
 "dst_host_srv_count",
 "dst_host_same_srv_rate",
 "dst_host_diff_srv_rate",
 "dst_host_same_src_port_rate",
 "dst_host_srv_diff_host_rate",
 "dst_host_serror_rate",
 "dst_host_srv_serror_rate",
 "dst_host_rerror_rate",
 "dst_host_srv_rerror_rate"]


# Ajout de la colonne 'target' à la liste des colonnes
columns.append('target')
print("nombre de colonnes sont : ",len(columns))

# Définition des types d'attaques
attacks_types = {
	'normal': 'normal',
'back': 'dos',
'buffer_overflow': 'u2r',
'ftp_write': 'r2l',
'guess_passwd': 'r2l',
'imap': 'r2l',
'ipsweep': 'probe',
'land': 'dos',
'loadmodule': 'u2r',
'multihop': 'r2l',
'neptune': 'dos',
'nmap': 'probe',
'perl': 'u2r',
'phf': 'r2l',
'pod': 'dos',
'portsweep': 'probe',
'rootkit': 'u2r',
'satan': 'probe',
'smurf': 'dos',
'spy': 'r2l',
'teardrop': 'dos',
'warezclient': 'r2l',
'warezmaster': 'r2l'
}
```

```python
# Chargement des données depuis le fichier CSV
path = "kddcup.data.gz"
df = pd.read_csv(path, names = columns)

# Suppression des doublons dans le jeu de données
df.drop_duplicates(subset=columns, keep='first', inplace = True)
print('Null values in dataset are',len(df[df.isnull().any(1)]))

# Création d'une nouvelle colonne 'Attack Type' en utilisant la correspondance avec 'target'
df['Attack Type'] = df.target.apply(lambda r:attacks_types[r[:-1]])
df.head()
```

```python
print(df.shape)
print(df.head())
df.isnull().sum()
num_cols = df._get_numeric_data().columns
```

```python
cate_cols = list(set(df.columns)-set(num_cols))
cate_cols.remove('target')
cate_cols.remove('Attack Type')

cate_cols
```

```python
# Suppression de colonnes contenant des valeurs NaN
df = df.dropna('columns')

# Sélection des colonnes avec plus d'une valeur unique
df = df[[col for col in df if df[col].nunique() > 1]]

df.sample(10)
```

```python
# Calcul de la matrice de corrélation
corr = df.corr()

# Affichage de la heatmap de corrélation
plt.figure(figsize =(15, 12))
sns.heatmap(corr)
plt.show()
```

```python
# Suppression de colonnes spécifiques du dataframe
df.drop('num_root', axis = 1, inplace = True)


df.drop('srv_serror_rate', axis = 1, inplace = True)


df.drop('srv_rerror_rate', axis = 1, inplace = True)


df.drop('dst_host_srv_serror_rate', axis = 1, inplace = True)


df.drop('dst_host_serror_rate', axis = 1, inplace = True)

df.drop('dst_host_rerror_rate', axis = 1, inplace = True)


df.drop('dst_host_srv_rerror_rate', axis = 1, inplace = True)


df.drop('dst_host_same_srv_rate', axis = 1, inplace = True)
```

```python
# Utilisez la méthode value_counts() pour compter chaque valeur dans la colonne "AttackType"
attack_type_counts = df['Attack Type'].value_counts()

# Calculez le nombre total d'attaques
total_attacks = len(df)

# Calculez les pourcentages
attack_type_percentages = (attack_type_counts / total_attacks) * 100

# Affichez les comptages
print("Comptage des attaques :")
print(attack_type_counts)
print("\nPourcentages des attaques :")
print(attack_type_percentages)

# Tracer les résultats en pourcentage
plt.figure(figsize=(10, 6))
attack_type_percentages.plot(kind='bar')
plt.title('Répartition des attaques en pourcentage')
plt.xlabel('Type d\'attaque')
plt.ylabel('Pourcentage')
plt.show()
```

```python
# Mapping des valeurs de la colonne 'protocol_type' à des entiers
pmap = {'icmp':0, 'tcp':1, 'udp':2}
df['protocol_type'] = df['protocol_type'].map(pmap)
```

```python
# Mapping des valeurs de la colonne 'flag' à des entiers
fmap = {'SF':0, 'S0':1, 'REJ':2, 'RSTR':3, 'RSTO':4, 'SH':5, 'S1':6, 'S2':7, 'RSTOS0':8, 'S3':9, 'OTH':10}
df['flag'] = df['flag'].map(fmap)

# Suppression de la colonne 'service'
df.drop('service', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Importation des modules pour la division des données et la mise à l'échelle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
# Séparation des données en variables explicatives (X) et la variable cible (y)
df = df.drop(['target', ], axis = 1)
print(df.shape)

y = df[['Attack Type']]
X = df.drop(['Attack Type', ], axis = 1)
```

```python
# Mise à l'échelle des variables explicatives
sc = MinMaxScaler()
X = sc.fit_transform(X)


# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
print('='*40)
print('='*40)

# Définition des types d'attaques sous forme numérique
attacks_types = {0: "normal", 1: "dos", 2: "u2r", 3: "r2l", 4: "probe"}
```

# Modèle 1 : Decision Tree Classifier

```python
# Modèle 1 : DecisionTreeClassifier
print("model 1 : DecisionTreeClassifier")
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix,classification_report
import time
from sklearn.tree import DecisionTreeClassifier

# Création d'un classificateur Decision Tree
clfd = DecisionTreeClassifier(criterion ="entropy", max_depth = 4)
```

```python
# Entraînement du modèle
start_time = time.time()
clfd.fit(X_train, y_train.values.ravel())
end_time = time.time()
print("Training time: ", end_time-start_time)
```

```python
# Prédiction sur les ensembles d'entraînement et de test
start_time = time.time()
y_test_pred = clfd.predict(X_test)
y_train_pred = clfd.predict(X_train)
end_time = time.time()
print("Testing time: ", end_time-start_time)

# Évaluation du modèle Decision Tree
print("Train score (accuarcy) is:", clfd.score(X_train, y_train))
print("Test score(accuarcy) is:", clfd.score(X_test, y_test))
```

```python
# Prédire les classes avec votre modèle
y_pred = clfd.predict(X_test)

# Calculer les metriques pour chaque classe
precision = precision_score(y_test, y_pred, average= None)
recall = recall_score(y_test, y_pred, average=None)
f1 = f1_score(y_test, y_pred, average=None)

# Vérifier le type de la variable precision
print(type(precision))
```

```python
# Convertir la variable precision en liste si elle est de type numpy array
if isinstance(precision, np.ndarray):
    precision = precision.tolist()
if isinstance(recall, np.ndarray):
    recall = recall.tolist()
if isinstance(f1, np.ndarray):
    f1 = f1.tolist()

# Boucler sur les classes d'attaque et afficher la précision correspondante
for attack_type, p, r, f in zip(attacks_types.values(), precision, recall, f1):
    print(f"Metrics for {attack_type} attacks: Precision={p:.6f}, Recall={r:.6f}, F1-Score={f:.6f}")
```

```python
# Calcul de l'exactitude (accuracy) pour l'ensemble d'entraînement et de test
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Calcul de la précision (precision) pour l'ensemble d'entraînement et de test
# Utilisation de l'option 'weighted' pour calculer la précision moyenne pondérée
train_precision = precision_score(y_train, y_train_pred,average='weighted')
test_precision = precision_score(y_test, y_test_pred,average='weighted')

# Calcul du rappel (recall) pour l'ensemble d'entraînement et de test
# Utilisation de l'option 'weighted' pour calculer le rappel moyen pondéré
train_recall = recall_score(y_train, y_train_pred,average='weighted')
test_recall = recall_score(y_test, y_test_pred,average='weighted')

# Calcul du score F1 (F1 score) pour l'ensemble d'entraînement et de test
# Utilisation de l'option 'weighted' pour calculer le score F1 moyen pondéré
train_f1 = f1_score(y_train, y_train_pred,average='weighted')
test_f1 = f1_score(y_test, y_test_pred,average='weighted')

# Affichage des métriques d'évaluation pour l'ensemble d'entraînement
print("Train precision is:", train_precision)
print("Train recall is:", train_recall)
print("Train F1 score is:", train_f1)

# Affichage des métriques d'évaluation pour l'ensemble de test
print("Test precision is:", test_precision)
print("Test recall is:", test_recall)
print("Test F1 score is:", test_f1)
```

```python
# Affichage de la matrice de confusion
C = confusion_matrix(y_test, y_test_pred)
cm_df = pd.DataFrame(C)
labels = ['dos', 'probe', 'r2l', 'u2r', 'normal']
plt.figure(figsize=(20,15))
sns.set(font_scale=1.4)
sns.heatmap(cm_df, annot=True, annot_kws={"size":12}, fmt='g',xticklabels=labels , yticklabels=labels )

plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
    
plt.show()
```

# Model 7 : Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
# Définition de la fonction qui crée et retourne le modèle
def fun():
    model = Sequential()
    
    # Ajout de la première couche Dense avec 30 neurones en sortie, input_dim spécifie la dimension de l'entrée
    model.add(Dense(31,input_dim =31,activation = 'relu',kernel_initializer='random_uniform'))
    
    # Dans la couche suivante, nous n'avons pas besoin de spécifier input_dim, car le modèle est séquentiel,
    # et la sortie de la couche précédente est l'entrée de celle-ci
    model.add(Dense(1,activation='sigmoid',kernel_initializer='random_uniform'))
    
    #5 classes-normal,dos,probe,r2l,u2r
    model.add(Dense(5,activation='softmax'))
    
    # Configuration du modèle avec une fonction de perte "categorical_crossentropy" pour des classes multiples,
    # l'optimiseur "adam" et la métrique d'évaluation "accuracy"    
    model.compile(loss ='categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
    
    return model
```

```python
# Création d'un classificateur Keras utilisant la fonction fun() définie ci-dessus
# Nous spécifions également le nombre d'époques (epochs) et la taille du lot (batch_size)
model7 = KerasClassifier(build_fn=fun,epochs=20,batch_size=64)
```

```python
# Mesure du temps de début
start = time.time()

# Entraînement du modèle avec les données d'entraînement X_train et les étiquettes y_train
model7.fit(X_train, y_train.values.ravel())

# Mesure du temps de fin
end = time.time()
```

```python
# Affichage du temps d'entraînement
print('Training time :',(end-start))
```

```python
# Mesure du temps de début
start_time = time.time()

# Prédiction des étiquettes pour les données d'entraînement X_train
Y_train_pred7 = model7.predict(X_train)

# Mesure du temps de fin
end_time = time.time()

# Affichage du temps de prédiction pour les données d'entraînement
print("Training time: ",end_time-start_time)
```

```python
# Mesure du temps de début
start_time = time.time()

# Prédiction des étiquettes pour les données de test X_test
Y_test_pred7 = model7.predict(X_test)

# Mesure du temps de fin
end_time = time.time()
```
